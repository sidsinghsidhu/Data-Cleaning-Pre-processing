```python
# Feature Engg. - Data Cleaning
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
df['Arrival_Hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_Min'] = df['Arrival_Time'].str.split(':').str[1]
df['Arrival_Hour'] = df['Arrival_Hour'].astype('int')
df['Arrival_Min'] = df['Arrival_Min'].astype('int')
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Define categorical columns (IMPORTANT: list of strings, not tuple)
categorical_cols = ['Airline', 'Source', 'Destination', 'Additional_Info']

# Create OneHotEncoder object
encoder = OneHotEncoder(sparse_output=False, drop='first')

# Fit and transform categorical columns
encoded = encoder.fit_transform(df[categorical_cols])

# Convert encoded array to DataFrame
encoded_df = pd.DataFrame(
    encoded,
    columns=encoder.get_feature_names_out(categorical_cols),
    index=df.index
)

# Drop original categorical columns
df_numeric = df.drop(columns=categorical_cols)

# Combine numeric data + encoded data
df_final = pd.concat([df_numeric, encoded_df], axis=1)

# Check result
print("Final shape:", df_final.shape)
print("\nData types:\n", df_final.dtypes)
```
